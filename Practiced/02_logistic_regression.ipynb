# 2. Logistic Regression

---

# What problem does it solve?

Logistic Regression is used for **classification problems**, primarily:

### Binary Classification

Examples:

* Spam vs Not Spam
* Fraud vs Not Fraud
* Disease vs No Disease
* Customer Churn vs No Churn
* Pass vs Fail

Unlike Linear Regression, the output is not a continuous value.

Instead, the model predicts a **probability**.

Example:

```text
Probability of Fraud = 0.87
```

Then we convert that probability into a class:

```text
If probability > 0.5 → Fraud
Else → Not Fraud
```

---

# Why can't we use Linear Regression for Classification?

Suppose we use Linear Regression for spam detection.

Predictions could be:

```text
-1.2
2.4
5.6
```

These are invalid probabilities because probabilities must lie between:

```text
0 and 1
```

We need a function that always produces outputs in that range.

That's where Logistic Regression comes in.

---

# Core Intuition

Linear Regression computes:

z=w_1x_1+w_2x_2+\cdots+w_nx_n+b

Then Logistic Regression passes this value through a Sigmoid Function.

The sigmoid converts any real number into a probability between 0 and 1.

---

# What is the Sigmoid Function?

The sigmoid function is:

\sigma(z)=\frac{1}{1+e^{-z}}

Properties:

### If z is very large

```text
Probability ≈ 1
```

### If z is very small

```text
Probability ≈ 0
```

### If z = 0

```text
Probability = 0.5
```

---

# Intuition Behind Sigmoid

Think of sigmoid as a translator.

Before sigmoid:

```text
-100
50
300
```

After sigmoid:

```text
0.00
0.99
1.00
```

Now the output can be interpreted as probability.

---

# What is the model actually learning?

The model learns coefficients:

```text
Age
Income
Credit Score
```

and determines:

```text
How much does each feature increase
or decrease the probability
of belonging to a class?
```

---

# What are Odds?

Very common interview topic.

Suppose:

```text
Probability of winning = 0.8
```

Odds are:

```text
0.8 / 0.2 = 4
```

Meaning:

```text
4 times more likely to win than lose
```

---

# What are Log Odds?

Logistic Regression actually models:

```text
log(odds)
```

instead of probability directly.

Reason:

* Log odds can take any value
* Linear models work well on unbounded values

The sigmoid converts them back into probabilities.

---

# Cost Function

In Linear Regression we used MSE.

For Logistic Regression we use:

### Log Loss / Binary Cross Entropy

Why not MSE?

Because:

* The problem is classification.
* Sigmoid creates a nonlinear relationship.
* MSE leads to difficult optimization.

Cross Entropy works much better.

---

# Binary Cross Entropy Intuition

The model gets punished heavily when:

```text
Actual = 1
Predicted = 0.01
```

because it was extremely confident and wrong.

But gets rewarded when:

```text
Actual = 1
Predicted = 0.99
```

---

# Assumptions of Logistic Regression

Interviewers love this.

---

## 1. Binary Outcome

Target should be categorical.

Examples:

```text
0/1
Yes/No
Spam/Not Spam
```

---

## 2. Independent Observations

One observation should not depend on another.

---

## 3. Low Multicollinearity

Features should not be highly correlated.

Same issue as Linear Regression.

---

## 4. Linear Relationship with Log Odds

Important.

Not:

```text
Features ↔ Probability
```

Instead:

```text
Features ↔ Log Odds
```

must be approximately linear.

---

# Decision Boundary

After probability prediction:

```text
0.0 to 1.0
```

we choose a threshold.

Most common:

```text
0.5
```

Decision:

```text
Probability > 0.5 → Class 1
Probability ≤ 0.5 → Class 0
```

This threshold can be changed.

---

Suppose:

```text

10 Fraud
```

Model predicts:

```text

```

Accuracy:

```text

```

Looks amazing.

Actually useless.

This is why classification uses:

* Precision
* Recall
* F1 Score
* ROC-AUC


---

# F1 Score

Balances:


---

# ROC Curve

Shows:

```text
True Positive Rate

```

for different thresholds.

---

# AUC

Interpretation:

```text

```

---

# Bias vs Variance

Logistic Regression generally has:

```text
Higher Bias
Lower Variance
```

Similar to Linear Regression.

Because it is a relatively simple model.

---

# Advantages

### Easy to Interpret

Coefficients explain feature effects.

---

### Fast Training

Works well on large datasets.

---

### Probability Output

Provides confidence scores.

---

### Strong Baseline Model

Often the first model used in industry.

---

# Limitations

### Cannot Capture Complex Nonlinear Patterns

Decision Trees and Boosting often perform better.

---

### Sensitive to Outliers

Outliers can distort coefficients.

---

### Requires Feature Engineering

Complex interactions may need manual creation.

---

# When should you use Logistic Regression?

Use when:

✅ Binary classification

✅ Need explainability

✅ Small to medium datasets

✅ Fast baseline needed

✅ Probability estimates required

---

# When should you NOT use it?

Avoid when:

❌ Complex nonlinear patterns

❌ Image classification

❌ Highly complex feature interactions

❌ Extremely high-dimensional nonlinear problems

Then consider:

* Random Forest
* XGBoost
* Neural Networks

# Common Interview Questions

### Q1: Why is it called Logistic Regression if it does classification?

Because the model performs linear regression on the log odds and then applies a logistic (sigmoid) function to produce probabilities.

---

### Q2: Why do we use Sigmoid?

To convert any real-valued number into a probability between 0 and 1.

---

### Q3: Why not use Linear Regression for classification?

Linear Regression can predict values outside the range [0,1], making them invalid probabilities.

---

### Q4: What is Log Loss?

A loss function that penalizes incorrect and overconfident predictions.

---

### Q5: What is the difference between probability and odds?

Probability:

```text
P
```

Odds:

```text
P / (1-P)
```

---

### Q6: What happens if classes are highly imbalanced?

Accuracy becomes misleading.

Use:

* Precision
* Recall
* F1
* ROC-AUC

---

### Q7: What does a coefficient mean in Logistic Regression?

It represents the change in log odds associated with a one-unit increase in the feature.

---

### Q8: Can Logistic Regression handle multiclass classification?

Yes.

Using:

* One-vs-Rest (OvR)
* Multinomial Logistic Regression

# Confusion Matrix

Almost every classification metric comes from the Confusion Matrix.

Suppose we are building a **fraud detection system**.

Actual values:

```text
1 = Fraud
0 = Not Fraud
```

Predictions:

```text
1 = Fraud
0 = Not Fraud
```

After testing, we get:

|                     | Actual Fraud | Actual Not Fraud |
| ------------------- | ------------ | ---------------- |
| Predicted Fraud     | 40           | 10               |
| Predicted Not Fraud | 20           | 930              |

This table is called a **Confusion Matrix**.

---

## True Positive (TP)

Model predicted Fraud.

Actual was Fraud.

```text
TP = 40
```

Correct positive prediction.

---

## True Negative (TN)

Model predicted Not Fraud.

Actual was Not Fraud.

```text
TN = 930
```

Correct negative prediction.

---

## False Positive (FP)

Model predicted Fraud.

Actual was Not Fraud.

```text
FP = 10
```

Also called:

```text
Type I Error
False Alarm
```

Example:

A legitimate transaction gets blocked.

---

## False Negative (FN)

Model predicted Not Fraud.

Actual was Fraud.

```text
FN = 20
```

Also called:

```text
Type II Error
Missed Detection
```

Example:

Fraud transaction goes unnoticed.

---

# 1. Accuracy

## Definition

Percentage of correct predictions.

Formula:

```text
Accuracy =
(TP + TN)
/
(TP + TN + FP + FN)
```

Using our numbers:

```text
Accuracy =
(40 + 930)
/
1000
=
97%
```

---

## Interpretation

The model is correct 97% of the time.

Looks excellent.

But wait.

---

## Problem with Accuracy

Suppose:

```text
990 Non-Fraud
10 Fraud
```

Model predicts:

```text
Everything = Non-Fraud
```

Confusion Matrix:

|           | Fraud | Not Fraud |
| --------- | ----- | --------- |
| Fraud     | 0     | 0         |
| Not Fraud | 10    | 990       |

Accuracy:

```text
990 / 1000
=
99%
```

Amazing accuracy.

Terrible fraud detector.

Because:

```text
Detected Fraud = 0
```

This is why accuracy alone is dangerous.

---

# 2. Precision

## Question Precision Answers

> When the model predicts Positive, how often is it correct?

Formula:

```text
Precision =
TP
/
(TP + FP)
```

Using our numbers:

```text
40 / (40 + 10)
=
80%
```

---

## Interpretation

Out of all fraud alerts:

```text
50 alerts generated
40 actually fraud
```

Precision:

```text
80%
```

Meaning:

```text
20% false alarms
```

---

Email Spam Filter

Suppose:

```text
100 emails marked spam
```

Actual spam:

```text
95
```

Precision:

```text
95%
```

Only 5 legitimate emails were wrongly classified.

Good.

---

## When Precision Matters

When false positives are expensive.

Examples:

### Spam Detection

Legitimate email marked as spam.

Bad.

Bad.

### Police Facial Recognition

Bad.

Formula:

```text
/
```

Using our numbers:

```text
=
66.7%
```

---

## Interpretation

```text
60
```

Detected:

```text
40
```

```text

```

Recall:

```text
66.7%
```

```text

```

Only 5 patients missed.

Good.

# Precision vs Recall

This is one of the most common interview questions.

---

## High Precision, Low Recall

Model only predicts positive when extremely confident.

Example:

```text
Fraud predicted = 20
All 20 are truly fraud.
```

Precision:

```text
100%
```

Recall:

```text
Low
```

Many frauds missed.

---

## High Recall, Low Precision

Model predicts fraud very aggressively.

Example:

```text
Detects almost all fraud.
```

But flags many normal transactions too.

Recall:

```text
High
```

Precision:

```text
Low
```

---

# 4. F1 Score

## Why do we need F1?

Sometimes:

```text
Precision = 100%
Recall = 20%
```

Looks good.

But model misses most positives.

F1 balances both.

---

## Formula

F1 is the harmonic mean:

F1=\frac{2\times Precision\times Recall}{Precision+Recall}

---

## Example

Suppose:

```text
Precision = 80%
Recall = 60%
```

F1:

```text
≈ 68.6%
```

---

## Why Harmonic Mean?

Because it penalizes imbalance.

Example:

```text
Precision = 100%
Recall = 0%
```

F1:

```text
0
```

Which makes sense.

---

## When F1 Matters

Class imbalance.

Examples:

* Fraud detection
* Medical diagnosis
* Intrusion detection

---

# 5. Specificity

Often asked with Recall.

---

## Question Specificity Answers

> Out of all actual negatives, how many did we correctly identify?

Formula:

```text
Specificity =
TN
/
(TN + FP)
```

Using our numbers:

```text
930 / (930 + 10)
=
98.9%
```

---

## Interpretation

The model is excellent at recognizing normal transactions.

---

# Recall vs Specificity

Recall focuses on:

```text
Positive Class
```

Specificity focuses on:

```text
Negative Class
```

---

# 6. ROC Curve

ROC means:

```text
Receiver Operating Characteristic
```

It plots:

```text
True Positive Rate (Recall)
vs
False Positive Rate
```

for different thresholds.

---

## Why Different Thresholds?

Usually:

```text
0.5
```

is used.

But we could use:

```text
0.3
0.6
0.8
```

Each threshold changes:

* Precision
* Recall

ROC visualizes this tradeoff.

---

# 7. AUC

Area Under ROC Curve.

Measures overall separability.

---

## Interpretation

```text
0.5 → Random guessing
```

```text
0.6 → Poor
```

```text
0.7 → Fair
```

```text
0.8 → Good
```

```text
0.9+ → Excellent
```

---

## Interview Interpretation

AUC answers:

> How well can the model distinguish between positive and negative classes?

Higher is better.

---

# PR Curve (Advanced Interview)

Precision-Recall Curve.

Preferred when:

```text
Classes are highly imbalanced
```

Example:

Fraud Detection:

```text
99.9% normal
0.1% fraud
```

PR Curve is often more informative than ROC.
